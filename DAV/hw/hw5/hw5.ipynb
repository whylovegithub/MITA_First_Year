```python
df_selected = df[["country","year","sex","age","suicides_no","population","suicides/100k pop"," gdp_for_year ($) " 
,"gdp_per_capita ($)","generation"]]
df_selected.rename(columns={" gdp_for_year ($) ": "gdp", "gdp_per_capita ($)": "gdp_per_capita", "age":"age_group"})
df_selected2 = df_selected
```

```python
df_selected = df_selected[df_selected["year"]>=1990]
df_selected = df_selected[df_selected["year"]<=2015]
```

```python
df_selected.head()
```

```python
countryList = df_selected.country.unique()
yearList = df_selected.year.unique()
```

```python
my_list = []
for i in countryList:
    a = df_selected.loc[df_selected['country'] == i]
    z=0
    for c in a.year.unique():
        z = z+1
    print(z)
    if (z <=8):
        my_list.append(i)     
    z=0
```

```python
print(my_list)
```

```python
df_selected=df_selected.loc[~df_selected.country.isin(my_list)]
df_selected.query('@my_list not in country')
df_selected.head()
```

```python
df_selected.to_csv("master_1995.csv")
```
